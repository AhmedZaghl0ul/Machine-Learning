```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler


x = 2*np.random.rand(10000,1)
y = 1+x+np.random.randn(10000,1)
```

```python
stn = StandardScaler()
xscale = stn.fit_transform(x)
```

```python
xscale[:10]
```

```python
xtrain , xtest , ytrain , ytest = train_test_split(xscale ,y, test_size=0.2, random_state=44)
```

```python
mo = LinearRegression()
mo.fit(xtrain , ytrain)
ypred = mo.predict(xtest)
```

```python
mse = mean_squared_error(ytest , ypred)
print(mse)
```

```python
plt.scatter(xtrain, ytrain, color='blue', label='Training data')
plt.scatter(xtest, ytest, color='red', label='Testing data')
plt.plot(xtest, ypred, color='yellow', linewidth=2, label='Regression line (predicted)')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('Linear Regression Example')
plt.show()
```
