#**`Labeling`**

```python
cat = [i for i in df.columns if df[i].dtype == 'O']
numerical = [i for i in df.columns if df[i].dtype != 'O']
```

```python
from sklearn.preprocessing import LabelEncoder

lb = LabelEncoder()

for col in cat:
    df[col] = lb.fit_transform(df[col])
```

```python
x = df.drop('income' , axis=1)
y = df['income']
```

#**`Scaling`**

---


```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
xscaled = scaler.fit_transform(x)
xscaled = pd.DataFrame(xscaled , columns=x.columns)
```

#**`Feature Selection`**

---


```python
from sklearn.feature_selection import SelectKBest , chi2
selk = SelectKBest(chi2 , k=8)
xnew = selk.fit_transform(xscaled , y)
print(x.shape)
print(x.columns[selk.get_support()].to_list())
x = xnew
```

```python
x.shape
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import cross_val_score ,GridSearchCV , learning_curve , RandomizedSearchCV , StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier , AdaBoostClassifier , VotingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report , accuracy_score , roc_curve
from imblearn.over_sampling import SMOTE
```

#**`Split`**

---


```python
from sklearn.model_selection import train_test_split

xtrin , xtest , ytrain , ytest = train_test_split( x , y , random_state=42 , stratify=y , test_size=0.2)
```

```python
sm = SMOTE(random_state=42)
xtr , ytr = sm.fit_resample(xtrin , ytrain)
```

```python
from xgboost import XGBClassifier

xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)

param = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, 7, 10],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.7, 0.8, 1.0],
    'colsample_bytree': [0.7, 0.8, 1.0]
}

rn = RandomizedSearchCV(xgb, param_distributions=param, scoring='accuracy', cv=5, n_iter=5, n_jobs=-1, random_state=42)
rn.fit(xtr , ytr)
best_models = rn.best_estimator_
print(f"Best Params for XGB: {rn.best_params_}")

ypred = rn.predict(xtest)
print(classification_report(ytest , ypred))
```
