```python
x = df.drop('Grade' , axis=1)
y = df['Grade']
```

```python
x.shape , y.shape
```

```python
from sklearn.feature_selection import SelectKBest

sl = SelectKBest(k=21)
x_select = sl.fit_transform(x, y)
```

```python
from sklearn.preprocessing import MinMaxScaler

sc = MinMaxScaler()
x_scaler = sc.fit_transform(x_select)
```

```python
from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test = train_test_split(x_scaler , y , test_size=0.10 , random_state=42)
```

```python
from imblearn.over_sampling import RandomOverSampler

ros = RandomOverSampler(random_state=42)
xresampled, yresampled = ros.fit_resample(x_train, y_train)
```

```python
xresampled.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from xgboost import XGBClassifier
```

```python
param_grids = {
    'LogisticRegression': {
        'C': [0.001, 0.01, 0.1, 1, 10, 100],
        'penalty': ['l1', 'l2']
    },
    'SVC': {
        'C': [0.1, 1, 10],
        'kernel': ['linear', 'rbf']
    },
    'RandomForestClassifier': {
        'n_estimators': [100, 200, 500],
        'max_depth': [None, 10, 20]
    },
    'DecisionTreeClassifier': {
        'max_depth': [None, 10, 20],
        'min_samples_split': [2, 5, 10]
    },
    'KNeighborsClassifier': {
        'n_neighbors': [3, 5, 7, 9],
        'weights': ['uniform', 'distance']
    },
    'XGBClassifier': {
        'n_estimators': [100, 200],
        'learning_rate': [0.01, 0.1]
    }
}
```

```python
models = {}

models['LogisticRegression'] = LogisticRegression()
models['SVC'] = SVC()
models['RandomForestClassifier'] = RandomForestClassifier()
models['DecisionTreeClassifier'] = DecisionTreeClassifier()
models['KNeighborsClassifier'] = KNeighborsClassifier()
models['XGBClassifier'] = XGBClassifier()
```

```python
for model_name, model in models.items():
    print(f"Running GridSearchCV for {model_name}...")
    grid_search = GridSearchCV(model, param_grids[model_name], cv=5)
    grid_search.fit(xresampled, yresampled)
    print(f"Best parameters for {model_name}: {grid_search.best_params_}")
    print(f"Best cross-validation score for {model_name}: {grid_search.best_score_:.4f}")
    print("-" * 30)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

results = {}

for model_name, model in models.items():
    print(f"Evaluating {model_name} on the test set...")

    grid_search = GridSearchCV(model, param_grids[model_name], cv=5)
    grid_search.fit(xresampled, yresampled)

    best_model = grid_search.best_estimator_
    y_pred = best_model.predict(x_test)

    accuracy = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)

    print(f"Accuracy for {model_name}: {accuracy:.4f}")
    print(f"Classification Report for {model_name}:\n{report}")
    print(f"Confusion Matrix for {model_name}:\n{cm}")
    print("-" * 30)

    results[model_name] = accuracy

best_model_name = max(results, key=results.get)
print(f"\nBest performing model on the test set: {best_model_name} with Accuracy: {results[best_model_name]:.4f}")
```

```python
import pandas as pd

accuracy_df = pd.DataFrame(list(results.items()), columns=['Model', 'Accuracy'])
accuracy_df = accuracy_df.sort_values(by='Accuracy', ascending=False)

display(accuracy_df)
```
